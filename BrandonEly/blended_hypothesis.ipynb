```python
import numpy as np
import random
import statistics as stat
import matplotlib.pyplot as plt

# create initial population
pop = []
values = [0,1]
gen=0
for i in range(1000):
  pop.append(random.choice(values))

# create results list
res = [{'gen': gen, 'diversity': np.round(stat.stdev(pop),3)}]

# simulation
new_pop = []
for n in range(100): # number of gens to simulate
  pop_copy = new_pop.copy() # save copy to make next gen

  # make new population
  for offspring in range(1000): # pop size
    if gen == 0: # only instance when you have to use initial pop instead of pop_copy
      new_pop.append(stat.mean([random.choice(pop), random.choice(pop)]))
    else:
      new_pop.append(stat.mean([random.choice(pop_copy), random.choice(pop_copy)]))

  # add pop std dev to results
  res.append({'gen': gen, 'diversity': np.round(stat.stdev(new_pop),3)})
  gen+=1


# plot results
generations = [rec['gen'] for rec in res]
diversity = [rec['diversity'] for rec in res]
plt.plot(generations, diversity)
plt.xlabel('generation')
plt.ylabel('std dev')
plt.title('Blended traits simulation results')
plt.xticks(range(0, len(generations), 10))

plt.show()
```
