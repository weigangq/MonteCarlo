```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def blender(pop_size, gens ):
  fig, ax = plt.subplots(2, 1, figsize = (7, 10))
  rng = np.random.default_rng()
  pop = rng.binomial(1, .5, pop_size)
  avg = []
  avg.append(np.mean(pop))
  dev = []
  dev.append(np.std(pop))
  total = 0
  for gen in range(gens):
    new_pop = []
    for i in range(len(pop)):
      parents = np.random.choice(pop, 2, replace = False)
      birth = (parents[0] + parents[1]) / 2
      new_pop.append(birth)

    avg.append(np.mean(new_pop))
    dev.append(np.std(new_pop))
    pop = new_pop
  datas = {"Generation" : np.arange(0, gens+1), "Mean" : avg, "StdDev" : dev}
  dataframe = pd.DataFrame(datas)
  print(sns.lineplot(x = "Generation", y = "Mean", data = dataframe, ax = ax[0]). set(title = "Generation and Mean Relationship"))
  print(sns.lineplot(x = "Generation", y = "StdDev", data = dataframe, ax = ax[1]). set(title = "Generation and Standard Error Relationship"))
```

```python
blender(1000, 50)
```
